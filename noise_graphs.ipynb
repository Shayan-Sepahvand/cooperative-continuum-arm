```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gauss_error = np.load('./absolute_error/gauss_error.npy')
speckle_error = np.load('./absolute_error/speckle_error.npy')
salt_error = np.load('./absolute_error/salt_error.npy')
occlusion_error = np.load('./absolute_error/occlusion_error.npy')
dark_error = np.load('./absolute_error/dark_error.npy')
bright_error = np.load('./absolute_error/bright_error.npy')
```

```python
fig = plt.violinplot([gauss_error, speckle_error, salt_error, occlusion_error, bright_error, dark_error], showextrema = True, showmeans = True, showmedians = True)
custom_ticks = range(1, 7)
custom_labels = ['Gaussian', 'Speckle', 'Impulse', 'Occluded', 'Bright', 'Dark']  # Adjust labels as needed
plt.xticks(ticks = custom_ticks, labels = custom_labels)
fig['cmeans'].set_color('m')
fig['cmedians'].set_color('g')
axes = plt.gca()
axes.set_ylabel('MAE (rad)')
axes.set_xlabel('Group Number')
axes.grid(True)
for pc in fig['bodies']:
    pc.set_facecolor('yellow')
    pc.set_edgecolor('black')
    
plt.savefig('./Results_Paper/figures/violin_plot.eps', format='eps')
```
