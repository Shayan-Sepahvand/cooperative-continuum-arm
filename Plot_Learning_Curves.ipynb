Plot the training curves

```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

Load the json file

```python
file_path = './Training_History_Files/training_history_without_finetuning.json'
history_dict = json.load(open(file_path, 'r'))
```

Plot the learning curves

```python
font = {'weight' : 'bold',
        'size'   : 14}
plt.rc ('font', **font)


figure1 = plt.figure(figsize=(5,5))
RootMeanSquareError =np.array(history_dict['loss'])
SampleNo = np.arange(1,len(RootMeanSquareError) + 1)



plt.plot(SampleNo,RootMeanSquareError , color = 'blue')
plt.xlabel('Epochs' , fontsize = 14 , weight='bold')
plt.ylabel(r'MSE (rad$^2$)' , fontsize = 14 , weight='bold')
plt.savefig('./Results_For_Report/MSETransferLeraning.eps', format='eps')

figure2 = plt.figure(figsize=(5,5))
MeanAbsError = history_dict['mean_absolute_error']


plt.plot(SampleNo,MeanAbsError , color = 'blue')
plt.xlabel('Epochs' , fontsize = 14 , weight='bold')
plt.ylabel('MAE (rad)' , fontsize = 14 , weight='bold')
plt.savefig('./Results_For_Report/MAETransferLeraning.eps', format='eps')
```
